```python
# 1. ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°

import pandas as pd
import re

# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv('nsmc_train.csv', encoding = 'utf-8')

text_col = 'review'

# í…ìŠ¤íŠ¸ ì •ì œ í•¨ìˆ˜
def clean_text(s):
    # ë‹¨ì–´ ì†Œë¬¸ì í™”
    s = str(s).lower()
    # ë‘ë²ˆì§¸ í•œê¸€,ì˜ë¬¸,ìˆ«ì - ì •ê·œí‘œí˜„ì‹
    s = re.sub(r"[^ã„±-ã…ê°€-í£a-zA-Z0-9 ]", " ", s)
    # ì¤‘ê°„ì— ê³µë°±ì´ ì—¬ëŸ¬ ê³µë°±ì´ ìˆìœ¼ë©´ ê³µë°±ì„ í•˜ë‚˜ë§Œ ì ìš© / ì–‘ìª½ì˜ ê³µë°±ì„ ëª¨ë‘ ì œê±°
    return re.sub(r"\s+", " ",s).strip()

df = df.dropna(subset=[text_col]).copy()
df['cleaned'] = df[text_col].apply(clean_text)

print(df)
```

```python
# 2ë‹¨ê³„ - countervectorizer - ë‹¨ì–´ ë¹ˆë„ ë²¡í„°í™”

from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

# 1. ë²¡í„°ë¼ì´ì € ìƒì„± / ìƒìœ„ 1000ê°œ ë‹¨ì–´ê¹Œì§€ ìƒì„± / ìµœì†Œ ë‹¨ì–´ íšŸìˆ˜ê°€ 5ëŠ” ë„˜ì–´ì•¼ ì¶œë ¥
cv = CountVectorizer(max_features=1000, min_df = 5)

# 2. countervectorizerì ìš© - ë¬¸ì¥ë“¤ì„ ìˆ«ìë¡œ ë³€í™˜ (ë¬¸ì„œ-ë‹¨ì–´ í–‰ë ¬) -> í•™ìŠµ ì‹œí‚¤ê¸°
x_cnt = cv.fit_transform(df['cleaned'])
vocab = cv.get_feature_names_out()
```

```python
print(vocab)
```

```python
# 3. ë‹¨ì–´ë³„ ì „ì²´ ë¹ˆë„ ê³„ì‚°
# freq ë‹¨ì–´ë³„ë¡œ ëª‡ê°œì˜ ê°’ì´ ìˆëŠ”ì§€ë¥¼ íŒŒì•…í•  ìˆ˜ ìˆëŠ” ë¬¸ë²•
# axis = 0 -> ì—´ì— í•´ë‹¹í•˜ëŠ” ê³„ì‚°
# A1 -> 2ì°¨ì› ë°°ì—´ -> 1ì°¨ì›ìœ¼ë¡œ ë³€ê²½
freq = x_cnt.sum(axis=0).A1 # ì²«ë²ˆì§¸ ìˆëŠ”ê²ƒë§Œ ì¶œë ¥
print(freq)

```

```python
# argsort -> ì‚¬ìš©ì„ ìì£¼í•˜ëŠ” í–‰ë ¬ ì²˜ë¦¬ í•¨ìˆ˜
# ì‘ì€ ê°’ -> í°ê°’ ìˆœì„œë¡œ ì •ë ¬ëœ ì¸ë±ìŠ¤ ë°°ê²½ì„ ë°˜í™˜í•˜ëŠ” êµ¬ë¬¸
# freq = [3,1,3,1] -> np.argsort(freq) -> [1,3,0,2] -> # 1ë²ˆ ë‹¨ì–´, 3ë²ˆ ë‹¨ì–´, 0ë²ˆ ë‹¨ì–´, 2ë²ˆ ë‹¨ì–´
# [::-1] -> ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ë˜ì–´ì•¼ í•˜ëŠ”ê²ƒì„ -> ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ë°”ê¿”ì„œ ì§‘ê³„
# [:20] ìƒìœ„ 20ê°œì— í•´ë‹¹í•˜ëŠ” ìë£Œë¥¼ ì–»ê²Œ ë¨
top_idx = np.argsort(freq)[::-1][:20] # ìƒìœ„ 20ê°œë§Œ ì¸ë±ìŠ¤ ì¶œë ¥
print(top_idx)
```

```python
# top_idxì— í•´ë‹¹í•˜ëŠ” ë‹¨ì–´ëŠ” ë¬´ì—‡ì¸ì§€? ê°¯ìˆ˜ëŠ” ëª‡ê°œì¸ì§€?
top_terms = [(vocab[i], int(freq[i])) for i in top_idx]
```

```python
print(top_terms)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(max_features=1000, min_df=5)
X_tfidf = tv.fit_transform(df["cleaned"].head(5000))
vocab_t = tv.get_feature_names_out()

# ë‹¨ì–´ë³„ í‰ê·  TF-IDF ê³„ì‚°(ì¶”ê°€)
mean_tfidf = X_tfidf.mean(axis=0).A1
top_idx_t = np.argsort(mean_tfidf)[::-1][:20]
top_terms_tfidf = [(vocab_t[i], round(float(mean_tfidf[i]), 4)) for i in top_idx_t]
top_terms_tfidf
```
